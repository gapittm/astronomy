```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import sin, cos, pi, linspace
```

## Exercise 1

```python
# Variables
sizeFactor = 5
title = 'Exercise 1'
aspect = 'equal'
xAxisName = 'x axis'
yAxisName = 'y axis'
xAxisMin = -15
xAxisMax = 15
yAxisMin = -15
yAxisMax = 15
xOffset = 0
yOffset = 0

# Calculate size and location of the shapes
theta = linspace(0, 2 * pi)
x = sizeFactor * cos(theta) + xOffset
y = sizeFactor * sin(theta) + yOffset
x2 = sizeFactor * 2 * cos(theta) + xOffset
y2 = sizeFactor * 2 * sin(theta) + yOffset

# Set up figure
fig, ax = plt.subplots()
ax.set_title(title, color = 'C0')
ax.set_xlabel(xAxisName)
ax.set_ylabel(yAxisName)
ax.set_xlim(xAxisMin, xAxisMax)
ax.set_ylim(yAxisMin, yAxisMax)
ax.set_aspect(aspect)

# Plot shapes
ax.plot(x, y, 'C2')
ax.plot(x2, y2, 'C3', linestyle = '--')
ax.plot(x2, y, 'C4', linestyle = ':')

# Add additional bits
ax.arrow(-3,-3,1,1,head_width=1,head_length=1,)
ax.plot(0,0,'*',markersize=20,color='C1')
ax.text(0,-14,'Gregory Pittman')

# Save the plot
# plt.savefig('filepath')
```

## Exercise 2 (combined)

```python
def orbit(day, hour):
    # Size values
    axisLim = 10
    orbitRadius = 5
    planetRadius = 1

    # Location of planet
    theta = linspace(0, 2 * pi)
    orbitX = orbitRadius * cos(2 * pi * (day / 365))
    orbitY = orbitRadius * sin(2 * pi * (day / 365))
    
    # Shape of planet plus offset
    planetX = planetRadius * cos(theta) + orbitX
    planetY = planetRadius * sin(theta) + orbitY
    
    # Arrow direction and offset
    arrowX = planetRadius * cos(2 * pi * (hour / 24))
    arrowY = planetRadius * sin(2 * pi * (hour / 24))
    arrowPX = arrowX + orbitX
    arrowPY = arrowY + orbitY

    # Set up figure
    fig, ax = plt.subplots()
    ax.set_xlim(-axisLim, axisLim)
    ax.set_ylim(-axisLim, axisLim)
    ax.set_aspect('equal')

    # Plot shapes
    ax.plot(0, 0, '*', markersize = 20, color = 'C1')
    ax.plot(planetX, planetY)
    ax.arrow(arrowPX, arrowPY, arrowX, arrowY, head_width = 1, head_length = 1)
```

```python
# Day 0 at hour 12
orbit(0, 12)
```

```python
# Day 180 at hour 12
orbit(180, 12)
```
